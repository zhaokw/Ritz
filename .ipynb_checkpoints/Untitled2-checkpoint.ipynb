```python
# %load util.py
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('LLBean.xls', sheet_name='Sheet1')
df['A/F'] = df['Actual Demand'] / df['Forecast']

# Plot A/F Density Curve
plt.figure(figsize=(15,8))
sns.distplot(df['A/F'])
plt.title('A/F Density Curve')

# Plot A/F Cumulative-Distr Curve
plt.figure(figsize=(15,8))
sns.distplot(df['A/F'], hist_kws={'cumulative': True}, kde_kws={'cumulative': True})
plt.title('A/F Cumulative');
```
