```python
# %load util.py
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pandas.tseries.offsets import MonthBegin
import warnings
warnings.filterwarnings('ignore')

def plot3Sigma(sc: pd.DataFrame, col: str):
    df = sc[['Date2','Occupancy', col]]
    df['p'] = df[col]/df['Occupancy']
    mu = df[col].sum()/df['Occupancy'].sum()
    df['sigma'] = df.apply(lambda r: (mu*(1-mu)/r['Occupancy'])**.5 , axis=1)
    df['UCL'] = 3*df['sigma'] + mu
    df['LCL'] = -3*df['sigma'] + mu

    plt.figure(figsize = (12,6))
    plt.plot(df['Date2'],df['p'])
    plt.plot(df['Date2'],df['UCL'])
    plt.plot(df['Date2'],df['LCL'])
    plt.xlabel('Day')
    plt.ylabel('\'' + col+ '\' Complaint Ratio')
    plt.title(col + ' Day-by-Day P-Chart');
    
def plot3SigmaByMo(sc: pd.DataFrame, col: str):
    scm = sc.copy()
    scm['DM'] = pd.to_datetime(scm['Date2']) - MonthBegin(1)
    scm = scm.groupby('DM').agg('sum').reset_index()
    scm['p'] = scm[col]/scm['Occupancy']
    mu = scm[col].sum()/scm['Occupancy'].sum()
    scm['sigma'] = scm.apply(lambda r: (mu*(1-mu)/r['Occupancy'])**.5 , axis=1)

    plt.figure(figsize = (6, 4))
    plt.plot(scm['DM'],scm['p'])
    plt.plot(scm['DM'],3*scm['sigma']+scm['p'])
    plt.plot(scm['DM'],-3*scm['sigma']+scm['p'])
    plt.xlabel('Month')
    plt.ylabel(col + ' Complaint Ratio')
    plt.title(col + ' Month-by-Month P-Chart');
```

```python
xlsx = pd.ExcelFile('Ritz.xlsx') 
sheets = xlsx.sheet_names
dd, sc = pd.read_excel('Ritz.xlsx', sheet_name='Defect Data'), pd.read_excel('Ritz.xlsx', sheet_name='Summary Counts')
sc = sc.drop(len(sc)-1).fillna(int(0))
sc.head()
```

<h1>Step 1</h1>
<h2>3-Sigma Analysis for Each Service Parameter</h2>

```python
for i in range(3, len(sc.columns)-1):
    plot3Sigma(sc, sc.columns[i])
```

<h1>Step 2</h1>
<h3>Month-by-Month 3-Sigma Analysis</h3>

```python
for i in range(3, len(sc.columns)-1):
    plot3SigmaByMo(sc, sc.columns[i])
```

# Distribution of P-value for room cleaning
<br/> We see that the distribution of the complaint rate of room cleaning is highly right-skewed

```python
plt.figure()
sc['p_rm_clean'] = (sc['rm clean']/sc['Occupancy']).fillna(0)
plt.axvline(sc['p_rm_clean'].mean(), color='r', linestyle='--', label='mean')
sns.distplot(sc['p_rm_clean'], ax = plt.gca())
sc = sc.drop('p_rm_clean', axis = 1)

plt.title('Distribution of the complaint rate of room cleaning ')
plt.ylabel('Frequency')
plt.legend();
```

<h1>Step 3</h1>
<h3>Trend Analysis by Complaint Type in each Department</h3>

```python
from datetime import date
for i in range(3, len(sc.columns)-1):
    col = sc.columns[i]
    
    ddOrdered = dd[dd['Keyword']==col][['Date2','Department']]
    ddOrdered['DM'] = pd.to_datetime(ddOrdered['Date2']) - MonthBegin(1)
    ddOrdered = ddOrdered.groupby(['DM','Department']).\
                agg('count').unstack().fillna(0).stack().reset_index()
    
    deptCount = len(ddOrdered['Department'].value_counts())
    if deptCount < 10:
        plt.figure(figsize = (16,8) if len(ddOrdered['Department'].value_counts()) >= 5 else None)
        sns.pointplot(x='DM',y='Date2',hue='Department',data = ddOrdered)

        plt.gca().set_xticklabels([pd.to_datetime(dm).strftime('%Y-%m') for dm in ddOrdered['DM'].unique()])    
        plt.xlabel('Month')
        plt.ylabel('Complaints');
        plt.title('\'' + col + '\' Complaint by Department')
```
